## TOKENIZATION

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, DistilBertTokenizer, DistilBertForSequenceClassification
```

```python
model_name2 = "nlptown/bert-base-multilingual-uncased-sentiment"
```

```python
mymodel= AutoModelForSequenceClassification.from_pretrained(model_name2)
```

```python
mytokenizer2 = AutoTokenizer.from_pretrained(model_name2)
```

```python
classifier = pipeline('sentiment-analysis', model=mymodel, tokenizer=mytokenizer2)
```

```python
classifier('I am dissapointed in the movie')
```

```python
classifier('I am happy with the movie')
```

```python
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
```

```python
tokens = tokenizer.tokenize('I am happy with the movie')
tokens
```

```python
tokenizer.convert_tokens_to_ids(tokens)
```

```python
tokenizer.decode(tokenizer.convert_tokens_to_ids(tokens))
```
